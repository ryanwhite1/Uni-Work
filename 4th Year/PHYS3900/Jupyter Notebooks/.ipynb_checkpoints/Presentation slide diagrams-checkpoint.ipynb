```python
# star background image retrieved from https://www.eso.org/public/images/eso1837d/

%matplotlib inline
import lightkurve as lk
import matplotlib.pyplot as plt
import scipy
import numpy as np
import astropy.units as u
from lightkurve import search_targetpixelfile

import starry

import pymc3 as pm
import pymc3_ext as pmx
import exoplanet

starry.config.quiet = True
starry.config.lazy = True

lcflist = lk.search_lightcurve('DI Her')
spoc = lcflist[0].download()
```

```python
#create spherical star system presented in Gaudi et al. 2017
pri_map = starry.Map(udeg=2, gdeg=4) #ydeg = 2*order_approx udeg=2
pri_map[1] = 0.124
pri_map[2] = 0.262

pri = starry.Primary(pri_map, m=5.1, r=2.4,prot=1.07)
sec_map = starry.Map(udeg=2, gdeg=4)

sec_map[1] = 0.124
sec_map[2] = 0.262
sec_map.amp=0.99
sec = starry.kepler.Secondary(map=sec_map,
    m=4.4,  # mass in solar masses
    r=2.58,  # radius in solar radii
    porb=10.55004,  # orbital period in days
    inc=89.02,
    Omega=70,  # longitude of ascending node in degrees
    ecc=0.505,  # eccentricity
    w=326.6,  # longitude of pericenter in degrees
    t0=2018.6681,  # time of transit in days
)
sys = starry.System(pri, sec)

(spoc/np.median(spoc.flux.value)).plot()
time = spoc.time.value
model_flux = sys.flux(time).eval()
plt.plot(time, model_flux/np.median(model_flux));
```

```python
#primary limb darkening
u1_m = 0.124
u1_e = 0.1
u2_m = 0.262 
u2_e = 0.1


#primary mass and radius
A_m_m = 5.1
A_m_e = 0.2
A_r_m = 2.4
A_r_e = 0.3

#primary rot period and tpole
A_prot_m =  1.07 #in days
A_omega = 0.225 #dimensionless
A_prot_e = 0.1

A_tpole_m = 17300
A_tpole_e = 800

#secondary luminosity ratio
B_amp_l = 0.6
B_amp_u = 0.8
#secondary mass and radius
B_m_m = 4.4
B_m_e = 0.2
B_r_m = 2.5
B_r_e = 0.3

#secondary tpole
B_tpole_m = 15400
B_tpole_e = 800

#secondary vsini
B_vsini_m = 100
B_vsini_e = 30

#orbital parameters
orb_inc_m = 88.90036698
orb_inc_e = 1.0
orb_period = 10.54980498
orb_ecc = 0.50002396
long_periastron = 327.28588277
long_ascend = 90 #106


G_mks = 6.67e-11
Msun = 1.989e+30
Rsun = 6.95700e8
```

```python
import theano.tensor as tt

time = spoc.time.value
flux = spoc.flux.value/np.median(spoc.flux.value)
ferr = spoc.flux_err.value/np.median(spoc.flux.value)

with pm.Model() as model:
    # These are the variables we're solving for;
    # here we're placing wide Gaussian priors on them.
    BoundedNormal = pm.Bound(pm.Normal, lower=0)
    
    A_m = BoundedNormal("A_M", mu=A_m_m, sd=A_m_e, testval=A_m_m)
    A_r = BoundedNormal("A_R", mu=A_r_m, sd=A_r_e, testval=A_r_m)
    A_inc_rad = pmx.Periodic("A inc", lower=0, upper=np.pi/2)
    A_inc = pm.Deterministic("A inc deg", A_inc_rad*180/np.pi)
    A_prot = BoundedNormal("A_prot", mu=1.07, sd=0.1, testval=1.07)
    
    pm.Potential("isotropy", tt.log(tt.sin(A_inc_rad)))
    
    u1 = BoundedNormal("u1", mu=u1_m, sd=u1_e, testval=u1_m+0.001)
    u2 = BoundedNormal("u2", mu=u2_m, sd=u2_e, testval=u2_m-0.001)
    
    
    pri_map = starry.Map(udeg=2, ydeg=4)
    pri_map[1] = u1
    pri_map[2] = u2
    pri_map.inc= A_inc
    
    primary = starry.Primary(pri_map, m=A_m, r=A_r, prot=A_prot)
 
    B_r = BoundedNormal("B_R", mu=B_r_m, sd=B_r_e, testval=B_r_m)

    sec_map = starry.Map(udeg=2, ydeg=1)
    sec_map[1] = u1
    sec_map[2] = u2

    secondary = starry.kepler.Secondary(map=sec_map,
        m = 4.4,  # mass in solar masses
        r = B_r,  # radius in solar radii
        porb = orb_period, # orbital period in days
        inc = orb_inc_m,
#         Omega = 90,  # longitude of ascending node in degrees, assume pole on
        Omega = long_ascend, # from Liang et al (unless it's 90)
        ecc = orb_ecc,  # eccentricity
        w = long_periastron,  # longitude of pericenter in degrees
        t0=pm.Uniform("t0", lower=2018.4681, upper=2018.8681, testval=2018.6681),  # time of transit in days
    )
    
    system = starry.System(primary, secondary)
```

```python
with model:
    system.set_data(flux, C=ferr**2)

    # Prior on primary
    pri_mu = np.zeros(primary.map.Ny)
    pri_mu[0] = 1
    pri_L = np.zeros(primary.map.Ny)
    pri_L[0] = 1e-2
    pri_L[1:] = 1e-2
    primary.map.set_prior(mu=pri_mu, L=pri_L)
    
    # Prior on secondary
    sec_mu = np.zeros(secondary.map.Ny)
    sec_mu[0] = 0.9
    sec_L = np.zeros(secondary.map.Ny)
    sec_L[0] = 1e-4
    sec_L[1:] = 1e-4
    secondary.map.set_prior(mu=sec_mu, L=sec_L)


    pm.Potential("marginal", system.lnlike(t=time))
```

```python
with model:
    map_soln = pmx.optimize()
```

```python
with model:
    x = pmx.eval_in_model(system.solve(t=time)[0], point=map_soln)
    primary.map.amp = x[0]
    primary.map[1:, :] = x[1 : primary.map.Ny] / primary.map.amp
    secondary.map.amp = x[primary.map.Ny]
    secondary.map[1:, :] = x[primary.map.Ny + 1 :] / secondary.map.amp
    flux_model = pmx.eval_in_model(system.flux(t=time), point=map_soln)
map_soln
```

```python
def set_ax_white(ax):
    for side in ['bottom', 'top', 'right', 'left']:
        ax.spines[side].set_color('w')
    ax.tick_params(axis='x', colors='w')
    ax.tick_params(axis='y', colors='w')
    ax.yaxis.label.set_color('w')
    ax.xaxis.label.set_color('w')
    return ax
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(time-min(time), flux_model, c='w')
# ax.set_facecolor('k')

ax.set_ylabel('Brightness')
ax.set_xlabel('Time (days)')
ax = set_ax_white(ax)
fig.savefig('Presentation Diagrams\lightcurve.png', dpi=400, bbox_inches='tight', transparent=True)
```

```python
def cartesian_to_spherical(x, y, z):
    radius = np.sqrt(x**2 + y**2 + z**2)
    equat = np.arctan2(y, x)    #returns equatorial angle in radians, maps to [-pi, pi]
    polar = np.arccos(z / radius)
    polar = np.degrees(polar)
    equat = np.degrees(equat)
    # now need to shift the angles
    if np.size(equat) != 1:
        equat = np.array([360 - abs(val) if val < 0 else val for val in equat])  #this reflects negative angles about equat=180
    else:   #same as above, but for a single element. 
        equat = 360 - abs(equat) if equat < 0 else equat
    return (equat, polar, radius)
```

```python
# fig, ax = plt.subplots(figsize=(10, 10))
filenameSec = 'Presentation Diagrams/secondaryEclipse.png'
filenamePri = 'Presentation Diagrams/primaryEclipse.png'
filenameStan = 'Presentation Diagrams/standard.png'

with model:
    system.show(t=5.9+min(time), point=map_soln, figsize=(5, 5), file=filenameSec, cmap='cool_r') # secondary
    system.show(t=8.3+min(time), point=map_soln, figsize=(5, 5), file=filenamePri,  window_pad=2, cmap='Blues') #primary
    system.show(t=5.7+min(time), point=map_soln, figsize=(5, 5), file=filenameStan, window_pad=4) # misc
```

```python

plotPri_map = starry.Map(udeg=2, ydeg=4)
plotPri_map[1] = map_soln["u1"]
plotPri_map[2] = map_soln["u2"]
plotPri_map.inc = map_soln["A inc deg"]

plotPri = starry.Primary(plotPri_map, m=map_soln["A_M"], r=map_soln["A_R"], prot=map_soln["A_prot"])


plotSec = starry.kepler.Secondary(map=sec_map,
    m = 4.4,  # mass in solar masses
    r = B_r,  # radius in solar radii
    porb = orb_period, # orbital period in days
    inc = orb_inc_m,
    Omega = long_ascend, # from Liang et al (unless it's 90)
    ecc = orb_ecc,  # eccentricity
    w = long_periastron,  # longitude of pericenter in degrees
    t0=map_soln["t0"])
plotSys = starry.System(plotPri, plotSec)
```

```python
reducedTime = [t for i, t in enumerate(time) if i%5 == 0]
x, y, z = plotSys.position(reducedTime)
fig, ax = plt.subplots(figsize=(10, 10))
add = 1000
priX = x[0].eval() + add; priY = y[0].eval() + add; priZ = z[0].eval() + add
secX = x[1].eval() + add; secY = y[1].eval() + add; secZ = z[1].eval() + add
priEquat, priPol, priRad = cartesian_to_spherical(priX, priY, priZ)
secEquat, secPol, secRad = cartesian_to_spherical(secX, secY, secZ)
ax.plot(priEquat, priPol, c='w')
ax.plot(secEquat, secPol, c='#dbdbdb')
ax = set_ax_white(ax)
plt.axis('off')
fig.savefig('Presentation Diagrams/orbit.png', dpi=400, bbox_inches='tight', transparent=True)
```

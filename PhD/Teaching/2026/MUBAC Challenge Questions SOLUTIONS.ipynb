### 8. Checking if a number is odd or even
The cell below has some code already filled in. Write an if/else block to check if the number the user supplies in `number` is even or odd. If the number is even, print "Even."; conversely, if the number is odd, print "Odd."

_**hint**: how do we tell if a number is odd or even? note that every odd number is just an even number + 1, and even numbers are divisible by 2. You should use either modulo arithmetic (`%` in python) or integer division (`//`)._

```python
number = int(input("Enter a number: "))

if number%2 == 0:
    print("Even.")
else:
    print("Odd.")

# another way of doing it, using integer division, is below:
# if number//2 == number/2:
#     print("Even.")
# else:
#     print("Odd.")
```

### 9. if/elif/else
Using a single conditional block, print to the user if their input number is negative, exactly 0, between 0 and 20 (exclusive), or 'greater than or equal to' 20. 

```python
number = float(input("Enter a number: "))

if number < 0:
    print("Number is negative")
elif number == 0:
    print("Number is exactly 0.")
elif number < 20:
    print("Number is between 0 and 20.")
else:
    print("Number is greater than or equal to 20.")
```

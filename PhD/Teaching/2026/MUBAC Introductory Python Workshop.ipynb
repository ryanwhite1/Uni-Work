# The Print Function

One of the most useful things you can do in Python is 'print' something to a human who is running the program. Python has an inbuilt function to do this which we call by typing `print()` and putting whatever we want to print within the parentheses. 

To demonstrate this, I've made a Python cell below this block of text with some Python code already written inside. To run this, or any Python cell in a Jupyter notebook, we can click inside of the cell then `Shift+Enter` on our keyboard (or press the Run button in the toolbar above). Let's do that now!

```python
print("Hello all of you lovely MUBAC students.")
```

You can see that some text gets output below the code cell once we run it. This is where the output is displayed in Jupyter notebooks. 

When we want to print some text, we enclose it within quotation marks `""` or apostrophes `''`. You can see that the text within the parentheses above have turned a different colour - this indicates that it is a string of characters. 

You can also print other types of data, not just strings of characters. For example:

```python
print(2)       # this is us printing an integer (a whole number)
print(3.4)     # this is a floating point number, or a 'float' - a number with a decimal
```

You will note that I included some extra text in the above cell, prefaced with a hash `#` symbol. Python doesn't 'read' the text following after a hash, and so we can use this to include extra information to the _humans reading the code_ that the machine won't action. This is called commenting and you should always use comments to document the code you write!

There is quite a bit more functionality of the print function, though we won't get to all of it in this crash course. One feature you might find particularly useful is that we can print multiple different things at the same time by separating each piece of information with a comma `,`: 

```python
print("I am including", 3, 'pieces of information in this print call!')
```

You can see that each piece of information gets output to the same line with a space character ` ` separating them. Sometimes it might be useful to have a different separation character, and we can customise this by dictating the `sep=` keyword argument within the print function. Let's try again, this time separating each piece of information with a new line `'\n'`:

```python
print("Each string will", "be printed on", "a new line!", sep='\n')
```

# Variables and Data Types

There's a whole lot more to do in Python than just printing out predefined blocks of text. To get onto the juicier content, we'll first need to learn about variables and data types. 

In essence, a variable is just us giving a name to a piece of information within our code. We can then use this name to do other things with it, like printing, math, or more! We define a variable with an equals sign `=` like so:

```python
integer_variable = 5
```

You'll note that there was no output when we ran the above Python cell! This is because we weren't printing anything this time. With that said, the Python cell still ran successfully. Now I can use this variable, which I've called `integer_variable`, elsewhere in our code:

```python
print(integer_variable)
```

The above variable, we know, is a whole number; it's an integer. Within Python, this is its own data type. We can display the 'type' of data that this is by using the `type()` function like so:

```python
print(type(integer_variable))
```

This tells us that the variable `integer_variable` is an `'int'`, exactly as expected. 

There are a range of data types in Python -- more than I'll cover here -- but the most relevant ones for this workshop are:
 - `int` - Integers: whole numbers, e.g. `2`
 - `float` - Floating point numbers: numbers with a decimal place, e.g. `10.3`
 - `str` - String: a string of _any_ characters, enclosed within quotation marks or apostrophes, e.g. `"Hello"`
 - `bool` - Boolean: a logical data type that is either `True` or `False` (we'll talk about these later)
 
 You can store each of these data types within variables, like so:

```python
float_variable = -50.30934342
str_variable = "This is a string!"
bool_variable = True
```

To some extent, we can also take a variable of some data type and convert it to another type. For example:

```python
float_to_int = int(float_variable)
print(float_to_int)
```

We can see that when we convert our float variable, `float_variable`, to an integer that the decimal point gets shaved off and we're left with the whole number. 
We can also do the reverse:

```python
int_to_float = float(integer_variable)
print(int_to_float)
```

When we convert a whole number to a floating point number, we just tack a `.0` onto the number.

We can convert these numbers to strings easily, and, within reason, strings to numbers:

```python
int_to_str = str(integer_variable)
float_to_str = str(float_variable)

print("Integer is now of type", type(int_to_str), "with value", int_to_str)
print("Float is now of type", type(float_to_str), "with value", float_to_str)

numeric_str = "403421"
str_to_int = int(numeric_str)

print("String is now of type", type(str_to_int), "with value", str_to_int)
```

# All of Mathematics in 5 Minutes
### Or: "How I Learned to Stop Worrying and Love the ~~BOMDAS~~ PEMDAS

Python is a powerful tool to compute things (otherwise why would we be doing all of this stuff on a computer?). To compute things, we need to know how to do math! Python has all of your favourite mathematical operators, like
 - Addition, `+`
 - Subtraction, `-`
 - Multiplication, `*`
 - Division, `/`
 - Exponentiation, `**`
 - Modulo, `%`
 - Integer Division, `//`
 
The order of operations follows the PEMDAS acronym too: Parentheses -> Exponentiation -> Multiplication -> Division -> Addition -> Subtraction

We can also do math with variables themselves. Let's take a look at 3 examples in increasing complexity:

```python
# We can do math directly inside a print function:
print(2 + 5)

# but it's usually better to assign each calculation to its own variable:
calc_1 = 3 * 6 + 3
print(calc_1)

# even better would be to assign each of our parameters to their own variable, then do the calculation with these:
a = 10
b = 3
c = 2.3
d = 5

calc_2 = (a + b)**c / d
print(calc_2)
```

That's _pretty much_ it for math in Python! 

...except... 

We can also do (some) math with strings! For example, we can add strings together:

```python
str_1 = "First string"
str_2 = " goes before"
str_3 = ' the third string.'

full_str = str_1 + str_2 + str_3
print(full_str)
```

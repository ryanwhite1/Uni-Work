# Understanding Rows in a DataFrame

***A row*** represents ***one observation*** (one data record).

In the iris dataset, *each row corresponds to one iris flower with its measurements*.

You can check the number of rows and columns using: .shape

```python
print(f'Number of rows: {iris_df.shape[0]}, Number of columns: {iris_df.shape[1]}')
```

# Understanding the Index

Each ***row*** in a pandas DataFrame has an ***index***.

The index acts like *an identifier for each row*.

You can check the index of a dataframe using: .index

```python
iris_df.index
```

This means rows are labelled:

0, 1, 2, 3, ..., 148, 149

You can ***access a specific row*** using .iloc.

```python
# Example: get the first row.
print(iris_df.iloc[0])
```

```python
# Or the first five rows:
print(iris_df.iloc[0:5])
```

# Selecting Rows Using Conditions

Often we want to ***select rows that satisfy certain conditions***.

Example: select flowers with *sepal length greater than 5*.

```python
iris_df[iris_df["sepal_length"] > 5]
```

Explanation:

`df["sepal_length"] > 5` creates a True / False mask.

Pandas then ***keeps only rows where the condition is True***.



```python
# To confirm the inner brackets output
iris_df["sepal_length"] > 5 # It will create a True / False mask.
```

You can combine conditions using:



*   & (AND)

# Visualising Filtered Data

We can plot specific subsets of the data.

Let's have a look the Sepal Length Distribution of all flowers first.

```python
plt.hist(iris_df["sepal_length"])
plt.title("Sepal Length Distribution (All Flowers)")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

Now let's visualise **only setosa flowers**.

```python
plt.hist(setosa_df["sepal_length"])
plt.title("Sepal Length (Setosa Only)")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

**Comparing Results**

Think about the difference:

* First plot → all flowers

* Second plot → only setosa

Filtering the data allows us to examine specific groups more clearly.

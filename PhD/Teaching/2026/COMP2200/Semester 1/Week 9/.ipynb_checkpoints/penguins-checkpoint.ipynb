```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.dummy import DummyClassifier
from sklearn.impute import SimpleImputer
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier 
```

```python
# Penguins with imputation and scaling
penguins = pd.read_csv('penguins.csv')
X = penguins.drop(columns=['species'])
y = penguins['species']
```

```python
num_cols = [['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
cat_cols = [['island', 'sex']]
```

```python
num_cols = X.select_dtypes(include='number').columns
cat_cols = X.select_dtypes(exclude='number').columns
```

```python
num_pipe = Pipeline([
    ('impute', SimpleImputer()),
    ('scale', StandardScaler())
])
```

```python
cat_pipe = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocess = ColumnTransformer([
    ('num', num_pipe, num_cols),
    ('cat', cat_pipe, cat_cols)
])
```

```python
# peng_pipe = Pipeline([
#     ('prep', preprocess),
#     ('clf', LogisticRegression())
# ])
peng_pipe = Pipeline([
    ('prep', preprocess),
    ('clf', LogisticRegression())
])
```

```python
acc_scores = cross_val_score(peng_pipe, X, y)
print('Penguins accuracy:', acc_scores.mean())
# Fit the model and show strongest features
peng_pipe.fit(X, y)
```

```python
feature_names = peng_pipe.named_steps['prep'].get_feature_names_out()

for cls, coef in zip(peng_pipe.named_steps['clf'].classes_, peng_pipe.named_steps['clf'].coef_):
    top_pos = coef.argsort()[-10:][::-1]
    top_neg = coef.argsort()[:10]
    print(f"\nTop features for {cls}:")
    print('  Positive:')
    for i in top_pos:
        print(f"    {feature_names[i]}: {coef[i]:.3f}")
    print('  Negative:')
    for i in top_neg:
        print(f"    {feature_names[i]}: {coef[i]:.3f}")
```

```python
feature_names = peng_pipe.named_steps["prep"].get_feature_names_out()
print(feature_names)
print(peng_pipe['clf'].classes_)
print(peng_pipe['clf'].coef_)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

param_grid = {'C':np.logspace(-2, 3, 100)}
grid = Pipeline([
    ('prep', preprocess),
    ('clf', GridSearchCV(LogisticRegression(max_iter=100), param_grid=param_grid, cv=5, scoring='accuracy', refit=True))
])

grid.fit(X, y)
print(grid.named_steps['clf'].best_params_, grid.named_steps['clf'].best_score_)
best = grid.named_steps['clf'].best_estimator_
```

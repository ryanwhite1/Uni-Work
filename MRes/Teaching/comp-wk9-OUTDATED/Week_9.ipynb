## Part B2

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
engagement = pd.read_csv("student_engagement.csv")
features = ["prep_minutes", "dropin_visits", "practice_quiz_attempts", "quiz3_score"]
cluster_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("cluster", KMeans(n_clusters=3, random_state=2025))
])
engagement["cluster"] = cluster_pipeline.fit_predict(engagement[features])
centroids = cluster_pipeline.named_steps["scaler"].inverse_transform(
    cluster_pipeline.named_steps["cluster"].cluster_centers_
)
```

```python
colourmapping = {0:'tab:green', 1:'tab:blue', 2:'tab:red'}
```

```python
import matplotlib.lines as mlines

fig, ax = plt.subplots()
ax.scatter(engagement['prep_minutes'], engagement['quiz3_score'], c=[colourmapping[val] for val in engagement['cluster']])

handles = []
for cluster in range(3):
    handles.append(mlines.Line2D([], [], color=colourmapping[cluster], marker='o', linestyle='', label=cluster))
ax.legend(title='Cluster story', handles=handles, frameon=False)

ax.set(xlabel='prep_minutes', ylabel='quiz3_score')

for i, centroid in enumerate(centroids):
    ax.text(centroid[0], centroid[-1], f"<{i}>", horizontalalignment='center', verticalalignment='center')
```

```python
fig, ax = plt.subplots()
for cluster in range(3):
    engagement[engagement.cluster == cluster].plot.scatter(x='prep_minutes', y='quiz3_score', c=colourmapping[cluster], ax=ax, label=cluster)
for i, centroid in enumerate(centroids):
    ax.text(centroid[0], centroid[-1], f"<{i}>", horizontalalignment='center', verticalalignment='center')
ax.legend(title="Cluster story", frameon=False)
```

## Part B3

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate
```

```python
numeric_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])
preprocess = ColumnTransformer([("num", numeric_pipeline, features)])
clf = Pipeline([
    ("preprocess", preprocess),
    ("model", LogisticRegression(max_iter=500))
])
```

```python
X = engagement[[feature for feature in features]]
y = engagement['passed_quiz']
cv = cross_validate(clf, X, y, scoring=["accuracy", "f1_macro"])
cv
```

```python
clf.fit(X, y)
```

```python
quiz = pd.DataFrame({'actual':engagement['passed_quiz'],
                    'predicted':clf.predict(X)})
                    # 'probability':clf.predict_proba(X)})
```
